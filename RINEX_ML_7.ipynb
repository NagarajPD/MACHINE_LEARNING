```python
#step 4 -- is devide the data into input and output
#input should be always 2d
x= df.iloc[:,2:4].values
x
```

```python
x.ndim
```

```python
y=df.iloc[:,4].values
y
```

```python
#step 5 train and test variables - here we use train test split module
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
#variables name can change but their irder should not change
#when ever we consider tarin test split we have to consider 4 variables
#by default 75% of x and y goes to traning variables
#and rest 25% goes to testing variables
#random_state = 0 ,it is used to keep the variables constant training and testing variables constant
```

```python
print(x.shape) #(400,2) 100% of x
print(x_train.shape)  #(300,2) 75% of x
print(x_test.shape)  #(100,2)  25% of x
```

```python
print(y.shape) #(400,1) 100% of y
print(y_train.shape)  #(300,1)  75% of y
print(y_test.shape)   #(100,1) 25% of y
```

```python
#step 6 normalization or scaling(done only for inputs)
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#step 7 run a classifier or a regresser or clusterer
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#8 fit the model (maping or plotting of values in libraray)
model.fit(x_train,y_train) #we are creating our model using thr training variables
```

```python
#step 9 - predict the output or predictor variable
y_pred=model.predict(x_test) # we will predict the output using input testing value
```

```python
y_pred # preditted output values
```

```python
y_test #actual input
```

```python
#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
