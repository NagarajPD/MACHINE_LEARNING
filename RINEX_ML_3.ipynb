```python
#matplotlib libraray(graph and other data representation)
#used in stock market ,growth of a company, cricket match, fall of inflation
import matplotlib.pyplot as plt
```

```python
#line graph
x=[1,2,3,4]
y=[5,6,7,8]
#plt.plot(x-axis,y-axis)
plt.plot(x,y,c='Orange',marker='d')
plt.title('line graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
```

```python
#line graph
import numpy as np
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
plt.title('line graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.plot(a,b,c='orange',marker='d',linewidth='1',linestyle='dashed')
```

```python
#scatter plot
a=[1,2,3,4,5]
b=[2,3,4,5,6]
plt.scatter(a,b,c='Orange',marker='x')
plt.title('scatter plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#bar graph
name=['A','B','C','D','E']
num=[4,2,7,5,1]
plt.bar(name,num,color=['Lime','red','b','yellow','orange'])
plt.title('bar graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
#BAR GRAPH
name=['Raj','thomas','Mac','naresh']
weight=[55,86,78,65]
plt.barh(name,weight,color=['Lime','red','b','yellow'])
plt.title('bar graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
