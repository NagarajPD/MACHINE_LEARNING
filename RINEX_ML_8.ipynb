```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1 take the data and create dataframe
df=pd.read_csv('/content/IRIS.csv')
df
```

```python
#input and outpluts
#inputs=sepal length, sepal width, petal length, petal width
#outputs= species
```

```python
#step 4 diving into input and output
x=df.iloc[:,0:4].values
x
```

```python
x.ndim
```

```python
y=df.iloc[:,4].values
y
```

```python
y.ndim
```

```python
#5 train and test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
#by default test size is 0.25 but where we change into 0.2
#hence training -80% and testing varialbes -20%
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#step 6 not requered , why because my inputs are alreday scaled
```

```python
#step 7 #apply a classifies regresser or calssifier
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model
```

```python
#8 model fitting
model.fit(x_train,y_train)
```

```python
#9 predict the output or predictor variable
y_pred= model.predict(x_test)
y_pred #predicted variable
```

```python
y_test
```

```python
#step 10 accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
