```python
#1 take the data and create daatframe
#area considered in sq feet and prices in rs
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv')
df
#IMAGINARY STORY -Imagine a real estate broker comes to you and gives the below data set
#and says create a model which could predict the property prices based on the data i gave
```

```python
#we are not performing step 2, due to limited data
```

```python
#step 3- data visilation
import matplotlib.pyplot as plt
plt.scatter(df['Area'],df['Prices'])
```

```python
#DIVIDE THE DATA INTO INPUT AND OUTPUT
#INPUT - AREA COLUMN
#OUTPUT - PRICES COLUMN
#INPUT IS ALWAYS 2 DIMENSIONAL ARRAY
#OUTPUT IS ALWAYS 1 DIMENSIONAL ARRAY
```

```python
#in column space if i ever have a ':' then its a 2 dimensional array
#x=fd.iloc[row slicing ,column slicing]
x=df.iloc[0:6,0:1].values #.vales is used to covert column into array
#x=df.iloc[:,0:1].values---- (:)= will select all rows
#0:6 will consider index 0 to index 5
#0:1 will consider index 0 alone
x
```

```python
x.ndim
```

```python
#in column space if there is no ':' then its 1 dimnsional
y= df.iloc[:,1].values
y
```

```python
y.ndim
```

```python
#step 5 and 6 are not being performed due to limited data
```

```python
#7. RUN CLASSIFIER/REGRESSOR/CLUSTERER(APPLY SUITABLE ALGORITHM)
#sklearn.linear_model is a package(collection of libraries),LinearRegression-Libraray
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#step 8-fit the model(mapping / or ploting inputs with the output)
#LinearRegression.fit(x,y)
model.fit(x,y)
```

```python
#step 9 - predict the output (predictor variable)
y_pred=model.predict(x) #using the input values(x), we predict the output(y_pred)
y_pred #predicted output
```

```python
y #actual outputs
```

```python
#CONCLUSION;
#WE HAVE COMPARE Y_PRED AND Y. ON COMPARING WE FIND OUT THAT THERE IS A HUGE DIFFERENCE
#THIS HUGE DIFFERENCE DOESNT MEAN THAT MY MOEL IS PREDICTED WRONG
#IT ONLLY MEANS THAT MY MODEL IS NON LINEAR/LESS LINEAR

#LINEARITY OF A MODEL , DEPENDS ON THE NATURE OF THE DATA AND THE SIZE OF THE DATA AS WELL
```

```python
#INDIVIDUAL PREDICTION
model.predict([[1700]]) #now need to find the avg price for 1700 sqft
```

```python
#cross verfication technique
#y=mx+C- EQUATION OF A STRAIGHT LINE
#y - dependent variable I
#m - slope
#x - independent variable
#C - Y-intercept/Constant
```

```python
#TO FIND OUT m FROM OUR DATAFRAME
m=model.coef_
m
```

```python
C=model.intercept_
C
```

```python
#y=mx+C
m*1700+C
```
