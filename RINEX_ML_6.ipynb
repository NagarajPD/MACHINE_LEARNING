```python
# LINEAR REGRESSIION -MODEL 2(HERE WE CREATE OUR OWN DATASET)
import numpy as np
experience=np.arange(1,11)#it will print values from 1 to 10
np.random.seed(35)# wt ever random values are generated , we want those values to be constant
salary = np.random.randint(20000,100000,10)
salary=np.sort(salary)#it sorts all the values in ascending order
print(experience)
print(salary)
```

```python
#step 4 -divide the data into input and output
#input is always 2 d
x=experience.reshape(-1,1) #.reshape(-1,1) is used to convert 1d array to 2d array
x
```

```python
x.ndim
```

```python
y=salary
y
```

```python
y.ndim
```

```python
#step5 and step 6 not performing due to limited data
```

```python
#step number 7
#apply a classifier or regresser or clusterer
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#step 8 we have to fit the model #Maping inputs with outputs
#LinearRegression.fit(x,y)
model.fit(x,y)
```

```python
#9 predict the output/predictor variable
y_pred=model.predict(x)
y_pred
```

```python
y
```

```python
#INDIVIDUAL PREDICTION
model.predict([[2.5]]) #i want the predict the salary for 2.5 years
```

```python
#NOW LET US LEARN ABOUT STEP NO 6 - SCALING / NORMALIZATION
#These graphs are nowhere related to Machine Learning
#We are only understanding the concept of Normalisation or Scaling using these graphs
#normalization is done to increase accuracy
#BEFORE NORMALIZATION
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)#to keep rando,ly generated values constant
x1=np.random.randint(1,50,30)
x1=np.sort(x1)# it sorts in acsending order
x2=np.random.randint(10000,70000,30)
plt.plot(x1,x2)
```

```python
#2 AFTER NORMALIZATION
x1min= min(x1)
x1max=max(x1)
x2min= min(x2)
x2max=max(x2)
x1norm=(x1 - x1min )/( x1max - x1min)
x2norm=(x2 - x2min )/( x2max - x2min)
plt.plot(x1norm,x2norm)
```
